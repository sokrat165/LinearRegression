### Multicollinearity

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
# since our categorical data is not yet preprocessed, we will only take the numerical ones

variable=data[['Mileage','Year','EngineV']]
# we create a new data frame which will include all the VIFs
vif=pd.DataFrame()
vif['VIF']=[variance_inflation_factor(variable.values,i) for i in range (variable.shape[1])]
vif["Features"] = variable.columns
```

```python
vif
```

```python
# Since Year has the highest VIF, I will remove it from the model
data=data.drop(['Year','Price'],axis=1)
```

```python
data.info()
```

## Create dummy variables


```python
# It is extremely important that we drop one of the dummies, alternatively we will introduce multicollinearity
data=pd.get_dummies(data,drop_first=True)
```

```python
data.head(5)
```

```python
data.columns.values
```

```python
col=[ 'log_price','Mileage', 'EngineV','Brand_BMW',
       'Brand_Mercedes-Benz', 'Brand_Mitsubishi', 'Brand_Renault',
       'Brand_Toyota', 'Brand_Volkswagen', 'Body_hatch', 'Body_other',
       'Body_sedan', 'Body_vagon', 'Body_van', 'Engine Type_Gas',
       'Engine Type_Other', 'Engine Type_Petrol',  'Registration_yes']
```

```python
data=data[col]
data.head(5)
```

## Linear regression model

```python
target=data.log_price
inputs=data.drop('log_price',axis=1)
```

## scale

```python
scale=StandardScaler()
scale_fit=scale.fit(inputs)
scale_transform=scale.transform(inputs)
```

 ## train & test

```python
x_train,x_test,y_train,y_test=train_test_split(scale_transform,target,test_size=0.2,shuffle=True,random_state=42)
```

### Create the regression

```python
reg =LinearRegression()
reg.fit(x_train,y_train)
reg.predict(x_test)
```

```python
y_predict=reg.predict(x_train)
```

```python
plt.scatter(y_train,y_predict,alpha=0.2)
plt.xlim(6,13)
plt.ylim(6,13)
plt.show()
```

```python
sns.histplot(y_train-y_predict,kde=True,color='#C40C0C')
plt.title("Residuals PDF", size=18)
```

```python
reg.intercept_
```

```python
reg.coef_
```


## Testing


```python
y_predict_test=reg.predict(x_test)
```

```python
plt.scatter(y_test,y_predict_test,alpha=0.2)
```

```python
df_p=pd.DataFrame(np.exp(y_predict_test),columns=['prediction'])
```

```python
df_p.head()
```

```python
y_test=y_test.reset_index(drop=True)
df_p['actual']=np.exp(y_test)
```

```python
df_p
```

```python
df_p['Residual'] = df_p['actual'] - df_p['prediction']
df_p['Difference%'] = np.absolute(df_p['Residual']/df_p['actual']*100)
```
